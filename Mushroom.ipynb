## Data Cleaning

```python
# required imports

import pandas as pd
```

```python
mush = pd.read_csv("/home/tejsax/Downloads/mushroom.csv")
df = mush.copy()
mush
```

```python
def f(c):
    return ord(c)
```

```python
df["cap_shape"] = df["cap-shape"].apply(f)
del df["cap-shape"]

df["cap_surface"] = df["cap-surface"].apply(f)
del df["cap-surface"]

df["cap_color"] = df["cap-color"].apply(f)
del df["cap-color"]

df["bruises_"] = df["bruises"].apply(f)
del df["bruises"]

df["odor_"] = df["odor"].apply(f)
del df["odor"]

df["gill_attachment"] = df["gill-attachment"].apply(f)
del df["gill-attachment"]

df["gill_spacing"] = df["gill-spacing"].apply(f)
del df["gill-spacing"]

df["gill_size"] = df["gill-size"].apply(f)
del df["gill-size"]

df["gill_color"] = df["gill-color"].apply(f)
del df["gill-color"]

df
```

```python
df["stalk_shape"] = df["stalk-shape"].apply(f)
del df["stalk-shape"]

df["stalk_root"] = df["stalk-root"].apply(f)
del df["stalk-root"]

df["stalk_surfabove"] = df["stalk-surface-above-ring"].apply(f)
del df["stalk-surface-above-ring"]

df["stalk_surfbelow"] = df["stalk-surface-below-ring"].apply(f)
del df["stalk-surface-below-ring"]

df["stalk_colorabove"] = df["stalk-color-above-ring"].apply(f)
del df["stalk-color-above-ring"]

df["stalk_colorbelow"] = df["stalk-color-below-ring"].apply(f)
del df["stalk-color-below-ring"]

df["veil_type"] = df["veil-type"].apply(f)
del df["veil-type"]

df["veil_color"] = df["veil-color"].apply(f)
del df["veil-color"]

df
```

```python
df["ring_number"] = df["ring-number"].apply(f)
del df["ring-number"]

df["ring_type"] = df["ring-type"].apply(f)
del df["ring-type"]

df["spore_color"] = df["spore-print-color"].apply(f)
del df["spore-print-color"]

df["popu"] = df["population"].apply(f)
del df["population"]

df["habitat_"] = df["habitat"].apply(f)
del df["habitat"]

df
```

```python
def classf(s):
    if s == 'p':
        return 1
    else:
        return 0
    
df["Poisonous?"] = df["class"].apply(classf)
del df["class"]
df
```

```python
y = df["Poisonous?"]
y
```

```python
del df["Poisonous?"]
x = df
```

```python
x
```

```python
from sklearn.model_selection import train_test_split, cross_val_score, KFold, GridSearchCV

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.2)
x_test.shape
```

## Finding out the best classifier

### PCA

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit_transform(x_train)

tot = sum(pca.explained_variance_)
k = 0
curr_v = 0
while curr_v/tot < 0.75:
    curr_v += pca.explained_variance_[k]
    k += 1
k   
```

```python
pca = PCA(n_components=k) #changed the number of features from 23 to  6 xD
x_train_p = pca.fit_transform(x_train)
x_test_p = pca.transform(x_test)
x_train_p.shape
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(criterion="entropy", random_state=0)

grid = {"n_estimators":[5*i for i in range(1, 10)], "max_depth":[5*i for i in range (1, 10)]}
g_rf = GridSearchCV(rf, grid, cv=KFold(10, True, 0))
g_rf.fit(x_train_p, y_train)
g_rf.best_estimator_
```

```python
g_rf.best_score_
```

```python
clf = RandomForestClassifier(criterion='entropy', n_estimators=45, max_depth=25)
clf.fit(x_train_p, y_train)
clf.score(x_test_p, y_test)
```
